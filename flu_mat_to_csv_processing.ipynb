```python
import scipy.io
import numpy as np
import pandas as pd
from scipy.sparse import csc_matrix
```

```python
# load .mat file
mat_path = 'influenza_outbreak_dataset.mat'
data = scipy.io.loadmat(mat_path)
```

```python
# get features of flu_X_tr
flu_X_tr_raw = data['flu_X_tr']
city_keyword_matrix = []
for i in range(flu_X_tr_raw.shape[1]):
    matrix: csc_matrix = flu_X_tr_raw[0, i]
    summed = np.asarray(matrix.sum(axis=0)).flatten()
    city_keyword_matrix.append(summed)
```

```python
# transform to DataFrame
df = pd.DataFrame(city_keyword_matrix)
df.columns = [f'keyword_{i+1}' for i in range(df.shape[1])]
```

```python
# let flu_Y_tr be：As long as the city has an outbreak on one day out of 1095 days, it is recorded as 1;
# otherwise, it is 0
outbreak_sums = []
for i in range(flu_X_tr_raw.shape[1]):
    region_outbreak_daily = data['flu_Y_tr'][0, i]
    outbreak_sums.append(int(np.sum(region_outbreak_daily)))
```

```python
# save as a CSV file
df['outbreak'] = outbreak_sums
df.to_csv('flu_train_clean_final.csv', index=False)
```
