```python
import scipy.io
```

```python
import pandas as pd
```

```python
# open file
mat_path = "influenza_outbreak_dataset.mat"
data = scipy.io.loadmat(mat_path)

# Extract the keywords
flu_keywords = [str(x[0]) for x in data["flu_keywords"].flatten()]

# Construct the keyword mapping table
mapping_df = pd.DataFrame({
    "column": [f"keyword_{i+1}" for i in range(len(flu_keywords))],
    "name": flu_keywords
})

# Save as a CSV file
mapping_df.to_csv("keyword_mapping.csv", index=False)
print("The keyword mapping table has been saved as keyword_mapping.csv")
```
